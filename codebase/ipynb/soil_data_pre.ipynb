```python
import pandas as pd

df1 = pd.read_csv("/content/dataset1.csv")
df2 = pd.read_csv("/content/dataset2.csv")
```

```python
print(df1.columns)
print(df2.columns)
```

```python
df1 = df1.rename(columns={
    "Temperature": "temperature",
    "Humidity": "humidity",
    "Rainfall": "rainfall",
    "pH": "ph",
    "Crop": "label"
})
```

```python
columns = ["N","P","K","temperature","humidity","ph","rainfall","label"]

df1 = df1[columns]
df2 = df2[columns]
```

```python
coffee_df = df2[df2["label"] == "coffee"]
coffee_df.shape
```

```python
merged_df = pd.concat([df1, coffee_df], ignore_index=True)
merged_df = merged_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
merged_df["label"] = merged_df["label"].str.strip().str.title()
```

```python
merged_df["label"].value_counts()
```

```python
selected_crops = ["Coffee","Banana","Ginger","Pepper","Tapioca"]

merged_df = merged_df[merged_df["label"].isin(selected_crops)]
merged_df["label"].value_counts()
```

```python
merged_df.to_csv("final_soil_dataset.csv", index=False)
```

```python
merged_df = merged_df.sort_values(by="label").reset_index(drop=True)
numeric_cols = ["N","P","K","temperature","humidity","ph","rainfall"]

merged_df[numeric_cols] = merged_df[numeric_cols].round(2)
```
